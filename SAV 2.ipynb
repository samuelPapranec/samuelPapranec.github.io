#### Code 5: + Print Obvod Trojuholnika
• Black-Box Symbolic - Know Your Expected Output and Expected Inputs
• We always get from function only one(actual) output.<br>
• In function´s definition we put an expected inputs into ***()*** of function as *parameters* <br>
• When we call the function, we need to input required data into ***()*** as *positional arguments* <br>
• ***print()*** ´s output format = `<empty>`: output <br>

```python
#def - parameters a, b, c
def vypocitaj_obvod(a, b, c):
    print("Obvod trojuholnika je: ", a+b+c)

#call - arguments a, b, c
vypocitaj_obvod(1,2,3)
```

#### Code 6a: Out Full User Name<a id='01'> (dependency of 6b)</a> 
• A good practice in python is to use short names for function parameters.<br>
• A python string method ***strip()*** remove spaces at the beginning and at the end of the string.<br>
• ***return*** ´s output format = `<Out>`: output  -return does not output anything in general <br>
    

```python
def vypis_cele_meno(m, p):
    """Funkcia na vypis celeho mena pouzivatela, prve pismena velke, s osetrenim medzier
        m = meno 
        p = priezvisko
    """
    cele_meno = m.strip() + " " + p.strip()
    cele_meno = cele_meno.title()
#output
    return cele_meno
#call

"""Example of messed up spaces because of importing from file or database:"""
Gregor = "           marek  "
Krab = "   rybanský        "

vypis_cele_meno(Gregor, Krab)
#If we add:
#    print("Cele meno je: ",vypis_cele_meno(Krab, Gregor))
#Then the Out[num]: will change to print´s output
```

#### Code 6b:Print Full User Name + Orientation in Documentation of Function- (dependencies: [6a](#01))

```python
print("Cele meno je: ",vypis_cele_meno(Krab, Gregor))

print(vypis_cele_meno.__doc__)
```

#### Code 10 Module ***getpass***  - hide password when typing(simple login prompt)
• In Python version 3 and up input works with strings

```python
from getpass import getpass as gp
pouzivatel_meno = input("Zadajte meno pouzivatela: ")
heslo = gp("Zadajte heslo: ")

print(f"Vase Meno je: {pouzivatel_meno}, a Heslo: {heslo}")
```
